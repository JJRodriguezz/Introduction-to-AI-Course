**BAG OF WORDS**

Bag of Words (BoW) es una técnica en procesamiento de lenguaje natural (NLP) que representa un texto mediante la frecuencia de aparición de palabras sin tener en cuenta el orden. Básicamente, convierte el texto en un vector de números donde cada posición representa una palabra del vocabulario y el valor indica cuántas veces aparece en el documento.

BoW es útil cuando el orden de las palabras no es relevante para la tarea, como en análisis de sentimientos o clasificación de textos, donde lo importante es qué palabras están presentes y cuántas veces.

**IMPLEMENTACIÓN**

```python
import numpy as np


def get_bag_of_words(doc, vocab):
   word_count_dict = dict.fromkeys(vocab, 0)# Mapea cada palabra del vocabulario a una cuenta desde 0
   for word in doc: #recorremos el do a ver si aparece la palabra, si aparece sumamos en la clave
      if word in vocab: #Verificación de que la palabra este en el diccionario para evitar errores
        word_count_dict[word] += 1

   conversion = np.zeros(len(vocab)) #Inicialización de un vector de zeros con la longitud del vocabulario

   for i, word in enumerate(vocab):
       conversion[i] = word_count_dict[word] #a cada valor de vector le ajusto el numero de apariciones de la palabra
   return conversion


```

```python
import re
# Define tu vocabulario tan extenso y abstracto como deseees segun la funcionalidad
vocab = ["delicious", "tasty", "amazing", "happy", "satisfied", "breakfast", "eggs",
          "ham", "friends", "love", "like", "dislike", "enjoy", "great", "fantastic",
          "yummy", "perfect", "meal", "morning", "brunch", "scrambled", "boiled",
          "fried", "sunny", "side", "bacon", "toast", "sausage", "cheese", "fresh",
          "pepper", "salt", "hungry", "awesome", "excellent", "crispy", "savory",
          "comfort", "snack", "full", "crave", "greasy", "smooth", "protein",
         "recommend", "share", "gather", "memories", "happy", "flavor"'a', 'am',
        'could', 'dark', 'do', 'fox', 'goat', 'good', 'green',
        'ham', 'here', 'house', 'i', 'if', 'in', 'let', 'like', 'may', 'me',
        'that', 'the', 'them', 'there', 'they', 'train', 'tree', 'try', 'will',
        'with', 'would', 'you']


#En este caso se pasan los comnetarios como una lista de n posiciones donde cada uno sera convertido a su equivalente de BoW
doc = ["I am very glad to be here with them, They have been an important support, they are the eggs of my ham, thank for that",
       "This is amazing, I would like to apologize the authors of this, they are awesome ",
       "Really I do not like that kind of house because is so dark",
       "I suppost that they have a good stratgy fot eggs with ham, ¿what do you think?",
       "The scrambled eggs with ham were absolutely delicious!",
        "I love having breakfast with friends; it makes me so happy.",
        "This morning's brunch was fantastic, especially with crispy bacon and toast.",
        "I crave a tasty meal with eggs, cheese, and ham.",
        "There's nothing like a fresh, savory breakfast to start the day.",
        "Sharing delicious food with friends always makes for great memories.",
        "The fried eggs with a side of toast were amazing!",
        "Adding a bit of salt and pepper to scrambled eggs makes them perfect.",
        "Ham and cheese sandwiches are my favorite snack!",
        "I'm full and satisfied after that awesome breakfast with eggs and bacon."]
```

```python
bag_of_words=[]
for i in range(len(doc)):
  #PASAMOS CADA FRASE A MINUSCULA
  vec = doc[i].lower()
  #HACEMOS UNA DIVISIÓN POR CUALQUIER CARACTER NO ALPHANUMERIC (SIGNOS DE PUNTUACION O ESPACIOS)
  vec = re.split("\W", vec)
  # ELIMINACIÓN DE ESPACIOS PARA EL VECTOR DE BoW
  vec = [s for s in vec if len(s) > 0]
  bag_of_words.append(get_bag_of_words(vec, vocab)) #Añadimos vector a la lista de BoW que almacena todas las conversiones
```
